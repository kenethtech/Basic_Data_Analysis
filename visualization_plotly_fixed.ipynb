```python
from datasets import load_dataset
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = load_dataset("KamaleshwariThirumugam/Stock-AAPL")
```

```python
df = dataset['train'].to_pandas()
```

```python
df.info()
```

```python
df['Date']= pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.head()
```

```python
fig = px.bar(df, x='Date', y='Close', title='AAPL Stock Price')
fig.show()
```

```python
fig2 = px.line(df, x='Date', y=['Open', 'Close'], title='AAPL Stock Price')
fig2.show()
```

```python
fig3 = px.line(df, x='Date', y=['Volume'], title='AAPL Stock Volume')
fig3.show()
```

```python
fig4 = px.line(df, x='Date', y='Open', title='AAPL Stock Open')
fig4.show()
```

```python
fig5 = px.histogram(df, x='Date', y='Close', title='AAPL Stock Close')
fig5.show()
```

```python
fig6 = px.scatter_3d(df, x='Date', y='Close', z='Volume', title='AAPL Stock Close and Volume')
fig6.show()
```

```python
fig7 = px.scatter_3d(df, x='Date', y='High', z='Low', title='AAPL Stock High and Low')
fig7.show()
```

```python
fig8 = px.imshow(df.corr(), title='AAPL Stock Correlation')
fig8.show()
```
